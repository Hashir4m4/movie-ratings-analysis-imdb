```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style='whitegrid')
```

```python
df = pd.read_csv(r"C:\Users\abdul\movie-ratings-analysis\data\movies.csv")
df.head()
df.info()
df.describe(include='all')
```

```python
df.drop_duplicates(inplace=True)

# Year column
df['year'] = df['Released_Year']

# Genre - first genre
df['genre'] = df['Genre'].str.split(',').str[0]

# Convert IMDb rating to numeric
df['rating'] = pd.to_numeric(df['IMDB_Rating'], errors='coerce')

# Drop missing rating rows
df = df.dropna(subset=['rating'])
```

```python
sns.histplot(df['rating'], bins=20)
plt.title("Movie Ratings Distribution")
```

```python
df.groupby('genre')['rating'].mean().sort_values(ascending=False).plot(kind='bar')
plt.title("Average Rating by Genre")
```

```python
df['Runtime'] = df['Runtime'].str.replace(' min', '').astype(int)
```

```python
top_movies = df[df['No_of_Votes'] > 5000].sort_values(by='rating', ascending=False).head(10)
top_movies[['Series_Title','rating','No_of_Votes']]
```

```python
numeric_df = df.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(), annot=True)
```
